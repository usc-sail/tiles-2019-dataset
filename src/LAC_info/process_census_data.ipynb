```python
import argparse
import pandas as pd

from pprint import pprint
from pathlib import Path

from process_census_data import *
from process_deaths import *
```

```python
class Args:
    def __init__(self):
        self.path = Path("~/Desktop/").expanduser()
        self.cutoffs = range(0, 24)
        self.freq = "30T"
        self.shift = False

args = Args()
```

```python
unit = '4A'
```

```python
census = {}
deaths = {}
df = {}
for unit in ['4A', '4B']:
    census[unit] = process_unit(unit, args, save=False)
    deaths[unit] = get_deaths(args.path/'Mortalities_LAC+USC.xlsx', unit, cutoffs=args.cutoffs, freq=args.freq, shift=args.shift)
    df[unit] = census[unit].merge(deaths[unit], on=['datetime', 'unit'], how='outer').fillna(0)
    
results = pd.concat([df[unit] for unit in ['4A', '4B']]).reset_index(); results
```
